```python
import datetime

# Function to set a reminder
def set_reminder():
    try:
        reminder_text = input("Enter your reminder: ")
        reminder_date_str = input("Enter date for reminder (YYYY-MM-DD): ")
        reminder_time_str = input("Enter time for reminder (HH:MM): ")

        reminder_date = datetime.datetime.strptime(reminder_date_str, "%Y-%m-%d")
        reminder_time = datetime.datetime.strptime(reminder_time_str, "%H:%M").time()

        reminder_datetime = datetime.datetime.combine(reminder_date, reminder_time)

        # Get current date and time
        current_datetime = datetime.datetime.now()

        # Calculate time difference
        time_difference = reminder_datetime - current_datetime

        # Check if the reminder date is in the future
        if time_difference.total_seconds() > 0:
            print("Reminder set successfully.")
            return reminder_datetime, reminder_text
        else:
            print("Invalid reminder date/time. Please enter a future date/time.")
            return None, None
    except ValueError:
        print("Invalid input. Please enter the date and time in the specified format.")
        return None, None

# Main function
def main():
    reminders = []

    while True:
        print("\nREMINDER APPLICATION")
        print("1. Set Reminder")
        print("2. View Reminders")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            reminder_datetime, reminder_text = set_reminder()
            if reminder_datetime:
                reminders.append((reminder_datetime, reminder_text))
        elif choice == '2':
            print("\nREMINDERS:")
            for idx, reminder in enumerate(reminders, start=1):
                print(f"{idx}. {reminder[1]} - {reminder[0].strftime('%Y-%m-%d %H:%M')}")
        elif choice == '3':
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
