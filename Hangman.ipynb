```python
import random
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

logo = ''' 
 _                                             
| |                                            
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __  
| '_ \ / _` | '_ \ / _` | '_ ` _ \ / _` | '_ \ 
| | | | (_| | | | | (_| | | | | | | (_| | | | |
|_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                    __/ |                      
                   |___/    '''
lives = 6
wordlist = ["apple","orange","banana","Dog", "Cat", "Ball", "Fish", "Bird", "Tree", "House", "Sun", "Moon", "Book",
    "Chair", "Hat", "Boat", "Lamp", "Duck", "Frog", "Cake", "Star", "Bear", "Horse",
    "Apple", "Car", "Key", "Rain", "Snow", "Shoe", "Rose", "Cup", "Pen", "Door",
    "Table", "Fork", "Spoon", "Baby", "Fox", "Rabbit", "Bed", "Clock", "Guitar", "Flag",
    "Smile", "Orange", "Banana", "Elephant", "Butterfly", "Pizza", "Train", "Robot", "Tiger", "Mouse"]

print(logo)
chosen_word = random.choice(wordlist).lower()



display = []
for i in range(len(chosen_word)):
    display += "_"


end_game = False

while not end_game:   
    guess = input("Guess a letter: ").lower()
    
    if guess in display:
        print(f"You have already guess the letter:{guess}")
    for position in range(len(chosen_word)):

        letter = chosen_word[position]

        if letter == guess:
            display[position] = letter
    
    
    if guess not in chosen_word:
        print(f"You have guessed {guess},this letter is not in the word")
        lives-=1
        if lives == 0:
            end_game = True
            print("You lose! Game Over")
    
    
    print(f"{' '.join(display)}") 
        
    
    
    if "_" not in display:
        end_game = True
        print("You win! Game Over")
        
    print(stages[lives])
    

   
```
